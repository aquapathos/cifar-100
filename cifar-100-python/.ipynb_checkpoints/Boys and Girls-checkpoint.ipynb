```python
import numpy as np
from PIL import Image
import pandas as pd

def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

train = unpickle("train")
test = unpickle("test")
clabels = pd.read_csv("EJcoarse_label_names.txt", header=0)
flabels = pd.read_csv("EJfine_label_names.txt", header=0)
```

```python
import PIL.Image
import numpy as np
import math
def blob2img(blob):
    return PIL.Image.fromarray(np.dstack(blob))
    
def getimage(data,i):
    return blob2img(data[b'data'][i].reshape((3,32,32)))

def flat2image(flat):
    return blob2img(flat.reshape((3,32,32)))

def showimage(data, start=0):
    canvas = PIL.Image.new('RGB',(350,350),(255,255,255))
    dsize = len(data[list(data.keys())[0]])
    end = start + 100
    if start + 100 > dsize:
        end = dsize
    for i in range(10):
        for j in range(10):
            n = start+i*10+j
            if n >= end:
                break
            else:
                canvas.paste(getimage(train,start+i*10+j),(35*j,35*i))
    return canvas
```

```python
taxonomy = pd.read_csv("taxonomy.txt", header=0)
taxonomy = taxonomy.drop('id',axis=1)
taxonomy.head(8)
```

```python
# 名称からid を
def word2fcat(word):
    return taxonomy[taxonomy['fword']==word]['fcat'].iloc[0]
```

```python
word2fcat('カワウソ')
```

```python
def getCat(data,cat):
    flabels = data[b'fine_labels']
    images = data[b'data']
    extract = []
    for fl, img in zip(flabels,images):
        if fl == word2fcat(cat):
            extract.append(img)  
    return extract
```

```python
def showBimages(blobs, start=0):
    canvas = PIL.Image.new('RGB',(350,350),(255,255,255))
    dsize = len(blobs)
    end = start + 100
    if start + 100 > dsize:
        end = dsize
    for i in range(10):
        for j in range(10):
            n = start+i*10+j
            if n >= end:
                break
            else:
                canvas.paste(flat2image(blobs[start+i*10+j]),(35*j,35*i))
    return canvas

showBimages(getCat(test,'カワウソ'))
```

```python
blobsR,blobsG,blobsB = blobs[:,0,:,:], blobs[:,1,:,:], blobs[:,2,:,:]
mean = (np.average(blobsR),np.average(blobsG),np.average(blobsB))
blobsR,blobsG,blobsB = blobsR-mean[0], blobsG-mean[1],blobsB-mean[2]
blobs = blobs/255.0
blobs = np.array(blobs).astype(np.float32)
```

```python
import chainer
from chainer import Variable
import chainer.links as L
import chainer.functions as F
from chainer import optimizers
```

```python
class MyNet(chainer.Chain):
    def __init__(self):
        super(MyNet,self).__init__()
        with self.init_scope():
            self.conv1 = L.Convolution2D(3,64,3,1)
            self.conv2 = L.Convolution2D(64,128,3,1)
            self.l1 = L.Linear(None,128)
            self.l2 = L.Linear(None,2)

    def __call__(self,x,train=True):
        h = F.relu(self.conv1(x))
        h = F.max_pooling_2d(h,2,2)
        h = F.relu(self.conv2(h))
        h = F.max_pooling_2d(h,2,2)
        h = F.dropout(F.relu(self.l1(h)))
        if chainer.config.train:
            return self.l2(h)
        return F.softmax(self.l2(h))
```

```python
nn = MyNet()
```

```python
import chainer
from chainer import training
from chainer.datasets import tuple_dataset
from chainer.iterators import SerialIterator
from chainer.training import trainer, extensions
optimizer = optimizers.Adam()
optimizer.setup(nn)
train = tuple_dataset.TupleDataset(blobs, y)

train_iter = SerialIterator(train, batch_size=10, shuffle=True)
updater = training.StandardUpdater(train_iter, optimizer)
trainer = training.Trainer(updater, (20, 'epoch'), out='result')
```

```python
trainer.extend(extensions.Evaluator(train_iter, nn))

trainer.extend(extensions.LogReport())

trainer.extend(extensions.PrintReport(['epoch', 'main/accuracy', 'validation/main/accuracy']))

trainer.extend(extensions.ProgressBar())
trainer.run() 
```
